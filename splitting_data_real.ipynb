*Use the code block below to open `sales.csv` with python. Then create a variable called `content` which is equal to a list of each line of the file*

````python
#uses the with functon to open sales.csv and renames as sales
with open("sales.csv") as sales:
    #creates a variable named content and set it equal every line of sales
    content = sales.readlines()
    
````

*Using the `split` function and a `for` loop calculate the total sales in a month*

*Hint: you will need a seperate variable for each month*

````python
#i made a variable for every month
jan = 0
feb = 0
mar = 0
april = 0
may = 0
june = 0
july = 0
august = 0
sept = 0
oct = 0
nov = 0
dec = 0

#a for loop for content which was made earlier
for x in content:
    #splits x using , and sets it equal to y
    y = x.split(",")
    #prints out y
    print(y)
    #splits the 4th value in z by / and then sets it equal to z
    z = y[3].split("/")
    #prints out z
    print(z)
    #if statement which checks if z is equal to 1, 2, or 3 and then assign its approriate month
    if z[0] == '1':
        #adds the total value in the split y list and adds it to its approriate month.
        jan = jan + float(y[2])
    elif z[0] == "2":
        feb = feb + float(y[2])
    elif z[0] == '3':
        mar = mar + float(y[2])

#prints out all the months
print(jan)
print(feb)
print(mar)
````

*Using your calculated total sales create a graph to represent this data and write a written analysis*

````python
#imports matplotlib.pyplot
import matplotlib.pyplot as plt
#makes the plot and labels x axis with the months and y axis with the values
plt.plot(["January","February","March"],[jan,feb,mar])
#sets the minimum and maximum value
plt.ylim(0,70000)
#sets the title
plt.title("Monthly Earnings")
````

*Using the `split` function, seperate each row of data. Then append the customer's name to a list called `customers`*

````python
#creates a list with empty values
list = []
#creates a for loop using content which is the opened value of sales.csv
for x in content:
    #sets y to the split version of x using the parameter ,
    y = x.split(",")
    #appends the first value which is the name to list
    list.append(y[0])
#deletes customer from the list
del list[0]
#prints the list
print(list)
````
