```python
#imports everything i need
import pandas as pd
import matplotlib.pyplot as plt
#reads the csv and sets it as data
data = pd.read_csv("nypd.csv")
#uses the head method to print out everything into a table
data.head()
```

how does the crimes committed by 18-24 year olds average between males and females compare in different ethnicity?

Using the graph created below, it can be determined that the crimed committed by genders and race varies by quite a large amount. It can be inferred from the graph that males that are black commit the most crimes out of all the ethicnity in which male white hispanics then black hispanics follow closely. The disparity between can mostly be ignored except for black, white hispanic and black hispanic who take the lead considerably compared to other ethnicities.

How do crimes compare between genders of the same race?

Using the same graph, it can be reasonably inferred that the males usually committ more crimes than the females. Most of the time the disparity between genders isnt by a large number except for black and hispanics in which the disparity is greatly increased.

```python
#i made x equal to data but organized using groupby function and grouping by based on age group and race and then counting based on sex
filtered_df = data[data['AGE_GROUP'] == '18-24']
#sets x to a filtered version using perp race column and perp sex and the counts it
x = filtered_df.groupby(['PERP_RACE', 'PERP_SEX'])["AGE_GROUP"].count()
#sets the color of the bars in order of which i want it
bar_colors = ['tab:green','tab:green','tab:green','tab:green','tab:green','tab:green','tab:orange','tab:red','tab:green','tab:green','tab:orange','tab:green','tab:green','tab:green','tab:green','tab:green','tab:green','tab:green','tab:green','tab:red','tab:green']
#plots out the bar with the appropriate labels and color
x.plot(kind='bar',ylabel= "Number of crimes committed",xlabel="Gender and Race",color=bar_colors)
#titles the graph
plt.title("Crimes organized by gender and race")

```

how does number of crimes committed by male age group 18-24 compare to the number of crimes committed by males in the age group 25-44

Crimed committed by the male group in the ages 18-24 compared to the age group 25-44 greatly lower. How does the crimes committed in the age group of 25-44 compare between the genders?

Based on the graph it can be inferred that those in the age group of 25-44 are the ones most likely to commit a crime since both highest crimes committed for each gender reside in the age group of 25-44. It can also be seen however a great difference between females and males as males in the age group of 25-44 have almost about 3 times the crime number than that of females in the same age group.

```python
#makes a graph using the groupby function and groups by age group and sex and then counts based on age group.

#sets y to a organized version of data and counts
y = data.groupby(['AGE_GROUP','PERP_SEX'])['AGE_GROUP'].count()
#sets the bar colors in order
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange', 'tab:purple']
#plots the graph using the appropriate labels and colors
y.plot(kind='bar', ylabel="Numbers of crimed committed",xlabel="Age and gender",color=bar_colors)
#names the title of the graph
plt.title("crime organized by age group and specifically male")

```
